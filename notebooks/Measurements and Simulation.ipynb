# Measurements

```python
from vdi import extenderTest as ext 
from matplotlib import style
from skrf.constants import mil, inch
from skrf.media import RectangularWaveguide
from skrf.calibration import SOLT, UnknownThru,MRC


# dictionary of measured and ideal networks
all_ideals = rf.read_all('data/ideals/')
all_measured = rf.read_all('data/measured/')

switch_terms = (rf.Network('data/switch_terms/forward switch term.s1p'),
                rf.Network('data/switch_terms/reverse switch term.s1p'))

#  aligned lists of ideal/measured networks
ideals = [all_ideals[k] for k in ['st-st','pl-pl','es-qs','qs-es','xth']]
measured = [all_measured[k] for k in ['st-st','pl-pl','es-qs','qs-es','xth,0']]

solt = SOLT(ideals=ideals, measured=measured, name='12-term')

ut = UnknownThru(ideals=ideals, measured=measured, switch_terms=switch_terms,
                  name='UnknownThru')

# MRC requires the reflect standards to be ordered as SDDL
# and uses the radiating open (ro) in place of the load
mrc_ideals = [all_ideals[k] for k in ['st-st','es-qs','qs-es','ro-ro','xth']]
mrc_measured = [all_measured[k] for k in ['st-st','es-qs','qs-es','ro-ro','xth,0']]

mrc = MRC(ideals=mrc_ideals, measured=mrc_measured, switch_terms=switch_terms,
           name='MRC')

# create list of calibrations to loop through in comparison 
cals = [solt,ut,mrc]

#create rectangular waveguide media, to generate theoretical responses
freq = rf.Frequency(325,500,201,'ghz')
wg = RectangularWaveguide(frequency=freq, a=22*mil, z0 = 50, rho=1.8e-8, 
                          roughness=.1e-6) 

# generate some theroetical responses
ds = wg.delay_short(1*inch)
line = wg.line(1*inch)


def compare_cals(x):
    '''temp function to compare correcting a DUT with different calibrations'''
    outs=[]
    for k in cals:
        out = k.apply_cal(all_measured[x])
        out.name = k.name
        out = out['325-500ghz']
        outs.append(out)
    return rf.NetworkSet(outs)
```

# Simulation

```python
#### adjustable parameters ###
sigma_sij = .8 #transmission phase noise std dev (deg) 
sigma_sii = .2 #reflection phase noise std dev (deg)
C = 2e-15 # flange capacitance
L = 4e-10 # flange inductance
####                      ####

# load approximate error networks from touchstone files. These 
# were saved from a one of the calibrations above
X = rf.Network('data/error_ntwks/X.s2p')
Y = rf.Network('data/error_ntwks/Y.s2p')

# set the waveguide object's frequency to match our embedding networks
wg.frequency = X.frequency

def measure(ntwk):
    '''measures a network by cascading error networks X and Y,then adds noise'''
    out =  X**ntwk**Y
    out.name = ntwk.name
    out_noisy = out.copy()
    out_noisy.add_noise_polar(0,sigma_sij) # uncomment to model noise
    out.s[:,0,1] = out_noisy.s[:,0,1]
    out.s[:,1,0] = out_noisy.s[:,1,0]
    out.add_noise_polar(0,sigma_sii)
    return out

def delay_shorts(d1,d2):
    '''creates two-port network for pair of delay shorts of lengths d1,d2'''
    ds1 = wg.delay_short(d1,'deg')
    ds2 = wg.delay_short(d2,'deg')
    return rf.two_port_reflect(ds1,ds2)

# ideal standard definitions
ideals = [
    wg.short(nports=2, name='short'),
    delay_shorts(45,90),
    delay_shorts(90,45),
    wg.match(nports=2, name='match'),
    wg.thru(name='thru'),
    ]

# model for misalgined flange
def flange():
    return wg.shunt_capacitor(C)**wg.shunt_inductor(L)** wg.thru()

# actual (true) responses of calibration standards
actuals = [
    wg.short(nports=2, name='short'),
    flange()**delay_shorts(45,90)**flange(),
    flange()**delay_shorts(90,45)**flange(),
    wg.match(nports=2, name='match'),
    wg.thru(), # the central dowels were used during this measuremnt, so dont add 
    # flange misalignment
    ]


# create ficticious measurements from actuals
measured = [measure(k) for k in actuals]

# create calibrations
solt = rf.SOLT(
    ideals = ideals,
    measured = measured,
    name = 'SOLT'
    )

ut = rf.UnknownThru(
    ideals = ideals,
    measured = measured,
    name = 'Unknown Thru'
    )

mrc = rf.MRC(
    ideals = ideals,
    measured = measured,
    name = 'MRC'
    )

cals = [solt, ut, mrc ]

def cor(dut):
    '''temp function to correct a single dut with all calibrations'''
    ns= []
    for k in cals:
        c = k.apply_cal(measure(dut))
        c.name = k.name
        ns.append(c)
    return rf.NetworkSet(ns)
```
