# Reflectometer Calibration With Unknown Lossless Reflects

From a mathematical perspective, the problem of one-port calibration is the determination of a Mobius transformation by observing how a known set of points is altered by the transformation. It is well known that three points uniquely determine the parameters of the transform]. If more than three points are observed, it is possible to exchange the exact location of every point with different constraints. The SDDL solution employs two lossless delay shorts of unknown phase. This physical constraint implies that the short, and two delay shorts all lie on a common circle. While simple to visualize, enforcing this constraint algebraically leads to an involved solution. An simpler approach is desirable to ease implementation as well as to create a more extendable solution.

## 2.1 General Solution

The problem of calibration is generally solved in the language of scattering or transfer parameters but can equally well be solved in alternative domains such as impedance or admittance. The fact that the actual measurement hardware is based upon scattering parameters is irrelevant. Translating to and from different network representations amounts to a change of basis. For this formulation of SDDL we choose to work with impedance parameters because it makes the lossless constraint simple to enforce, ie $\Re(z)=0$
 . In effect, this transforms the constraint that points lay on a circle with the constraint that points are collinear.

Given  the true impedance of 4 one-port standards is $a,b,c,d$, then embedding them behind a two-port network produces the corresponding measurements  $a',b',c',d'$. These sets are related through the transform 

$$ x' = z_{11} - \frac{z_{12}z_{21}}{z_{22}+x} $$

```python
from skrf import  rand_c

# generate three random impedance parameters
z11,z22,z1221 = rf.rand_c(3)

# the input impedance of a load impedance, x, is given by
zin = lambda (x): z11-z1221/(z22+x) 
```

It is well known that the cross ratio, defined by 

$$cr(a,b,c,d) = \frac{a-b}{a-d}\frac{c-d}{c-b} $$

is an invariant of the transform, thus

$$\frac{a-b}{a-d}\frac{c-d}{c-b} = \frac{a'-b'}{a'-d'}\frac{c'-d'}{c'-b'}$$

```python
# cross ratio function
cr = lambda a,b,c,d: ((a-b)*(c-d))/((a-d)*(c-b)) 

# given 4 random complex numbers
a,b,c,d = rand_c(4)


# their corresponding input inmpedances are 
a_,b_,c_,d_ = [zin(k) for k in [a,b,c,d]]

# the cross ratio is invariant
abs(cr(a,b,c,d) -  cr(a_,b_,c_,d_)) < 1e-12
```

Because all measurements are observable, their cross ratio is known and  may be lumped into a single variable, 

$$\frac{a-b}{a-d}\frac{c-d}{c-b} =z $$

```python
z=cr(a_,b_,c_,d_)
z
```

The problem of SDDL is to determine the values of two  partially known standards, provided that these two standards are know to be lossless ( ie have zero real parts). Letting the lossless reflects be represented by $a$ and $b$, and re-calculating corresponding input impedances and their cross ratio.

```python
# a and b are lossless standards 
a,b = rand(2)*1j

# c and d  are fully known arbitrary standards
c,d = rand_c(2)

# then we have to re-calculate measure quantities and the cross ratio
a_,b_,c_,d_ = [zin(k) for k in [a,b,c,d]]
z = cr(a_,b_,c_,d_)
```

Now, given 

$$\frac{a-b}{a-d}\frac{c-d}{c-b} =z $$

Cross multiplying the above and grouping *known* terms, 


\begin{equation}
a\underbrace{\left(c-d-cz\right)}_{e}+b\underbrace{\left(d-c-dz\right)}_{f}+abz+\underbrace{cdz}_{g}=0\\
ae+bf+abz+g=0
\end{equation}


Solving this for $a$ yields, 

\begin{equation}
a=-\frac{fb+g}{zb+e}
\end{equation}


Given that $b$ is one dimensional, because its purely imaginary, this equation yields a circle. 



```python
e = c-d-c*z
f = d-c-d*z
g = c*d*z

# 'a' as a function of b 
a_func = lambda b: -(f*b+g)/(z*b+e)

# range of possible values for b 
b_vect = linspace(-1e3,1e3,5e4)*1j


# plot the function for a, for possible values of b
from skrf import plot_complex_rectangular as pcr
kw =dict(markersize=5,marker = 'o' ,ls='')
axvline(0,color='k')
axhline(0,color='k')

pcr(a_func(b_vect), label = r'$a=f(b)$')
axis('equal');
```

Now, we need to solve for the values of a that are purely imaginary, this corresponds to the intersection of the circle and the y-axis. Note we have the constraints

$$ a = -\bar{a}$$
$$ b = -\bar{b}$$

Plotting the first constraint yields two circles, reflected about the y-axis

```python
# plot the function for a, for possible values of b
axvline(0,color='k')
axhline(0,color='k')

pcr(a_func(b_vect), label = r'$a=f(b)$')
pcr(-conjugate(a_func(b_vect)), label = r'$-\bar{a}$')
axis('equal');
```

Using both constraints ($a = -\bar{a}$,  $ b = -\bar{b}$) and solving the equations for $b$

\begin{eqnarray}
-a & = & \bar{a}\nonumber \\
\frac{fb+g}{zb+e} & = & \frac{\bar{f}b-\bar{g}}{-\bar{z}b+\bar{e}}\nonumber \\
\left(fb+g\right)\left(-\bar{z}b+\bar{e}\right) & = & \left(\bar{f}b+\bar{g}\right)\left(zb+e\right)
\end{eqnarray}


expanding yields a quadratic in $b$ with scalar coefficients

\begin{eqnarray}
b^{2}\left(-f\bar{z}-\bar{f}z\right)+b\left(-g\bar{z}+\bar{g}z+f\bar{e}-\bar{f}e\right)
+\left(g\bar{e}+\bar{g}e\right) & = & 0\\
b^{2}\underbrace{\Re\left(-f\bar{z}\right)}_{A}+b\underbrace{j\Im\left(\bar{g}z+f\bar{e}\right)}_{B}+\underbrace{\Re\left(g\bar{e}\right)}_{C} & = & 0
\end{eqnarray}

Replacing the polynomial coefficients with $A,B,$ and $C$ and solvign for $b$

\begin{equation}
b=\frac{-B\pm\sqrt{B^{2}-4AC}}{2A}
\end{equation}


```python
A = -real(f*z.conj())
B = 1j*imag( f*e.conj() + g.conj()*z)
C = real( g*e.conj()) 

p = poly1d([A,B,C])
print(p)
```

The roots of this polynomial yield, possible two values of $b$

```python
roots = p.r
roots
```

```python
#which root is correct?
abs(b - roots)<1e-12
```

These have the corresponding values of for $a$

```python
possible_a = a_func(roots)
possible_a
```

```python
# plot the function for a, for possible values of b
axvline(0,color='k')
axhline(0,color='k')

pcr(a_func(b_vect), label = r'$a=f(b)$')
pcr(-conjugate(a_func(b_vect)), label = r'$-\bar{a}$')
pcr(possible_a,  label = 'possible values of $a$',**kw)
axis('equal');
```

## Special Case of Flush Short (SDDL )


It is instructive to compare the new formulation with the previously
derived solution, which assumes the availability of
a flush short. If one of the known standards is a short, then either
$c$ or $d$ will be 0, which by implies
that $g$ is 0. Reflecting on this condition geometrically, 
becomes, 

\begin{equation}
a=\frac{-fb}{zb+e}
\end{equation}

From this it is clear that when $b=0$, $a=0$ and thus, the right
hand side represents a circle that passes through the origin. Given that $a$ does not
equal 0, or else the cross ratio would collapse, $a$ must be the
other point on the circle which intersects the imaginary axis. This
is a unique solution, and thus no sign choice must be made to evaluate
$b$.



```python


# then we have to re-calculate measure quantities and the cross ratio
d= 0
a_,b_,c_,d_ = [zin(k) for k in [a,b,c,d]]
z = cr(a_,b_,c_,d_)
e = c-d-c*z
f = d-c-d*z
g = c*d*z

a_func = lambda b: -(f*b)/(z*b+e)

b_ = 1j*imag(f*e.conj())/real(f*z.conj())



axvline(0,color='k')
axhline(0,color='k')

pcr(a_func(b_vect), label = r'$a=f(b)$')
pcr(a_func(b_),  label = 'value of $a$',**kw)
axis('equal');
```

## Special Case SDDM

```python
# have to re-calculate measure quantities and the cross ratio
d = 0 
c = 1
a_,b_,c_,d_ = [zin(k) for k in [a,b,c,d]]
z = cr(a_,b_,c_,d_)

# the solution is directly found from
b_found = 1j*imag(z)/real(z) 

(abs(b_found -b )< rf.ALMOST_ZER0)
```

## Special Case SDDO

### Illustration of Numerical Instability

```python
# then we have to re-calculate measure quantities and the cross ratio
d = 0
c = 1e12 # numerical approximation to inf

def solution_given_c(c):
    a_,b_,c_,d_ = [zin(k) for k in [a,b,c,d]]
    z = cr(a_,b_,c_,d_)

    e = c-d-c*z
    f = d-c-d*z
    g = c*d*z

    A = -real(f*z.conj())
    B = 1j*imag( f*e.conj() + g.conj()*z)
    C = real( g*e.conj()) 

    p = poly1d([A,B,C])
    return p.r[0] # the second root is 0
```

```python
c_vect = logspace(10,20,101)
solutions = [solution_given_c(k) for k in c_vect]
loglog(c_vect, imag(solutions))
xlabel('Resistance Values of $c$ (Ohms)')
ylabel('Determined reactance for $b$ (Ohms)')
title('Illustration of Numerical Instability')
```

## Stable solution 

```python
# re-calculate measure quantities and the cross ratio
d = 0 
c = 1e12
a_,b_,c_,d_ = [zin(k) for k in [a,b,c,d]]
z = cr(a_,b_,c_,d_)
b_found = a-a*z

(abs(b_found -b )< rf.ALMOST_ZER0)
```
